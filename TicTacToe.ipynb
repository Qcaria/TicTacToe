```python
def printer(grid):
    print('   0'+'   '+ '1'+'   '+ '2')
    print('0  ' + grid[0][0]+' | '+grid[0][1]+' | '+grid[0][2])
    print('1  ' + grid[1][0]+' | '+grid[1][1]+' | '+grid[1][2])
    print('2  ' + grid[2][0]+' | '+grid[2][1]+' | '+grid[2][2])
```

```python
def evaluate_game(grid, player):
    end = True
    for i in range(3):
        row = list(grid[i][n] for n in range(3))
        column = list(grid[n][i] for n in range(3))
        
        if ('X' not in row or 'O' not in row) and ' ' not in row:
            if row[0] == player:
                return True, 1
            else:
                return True, -1
        elif ' ' in row:
            end = False
            
        if ('X' not in column or 'O' not in column) and ' ' not in column:
            if column[0] == player:
                return True, 1
            else:
                return True, -1
        elif ' ' in column:
            end = False
            
    diagon1 = list(grid[i][i] for i in range(3))
    diagon2 = list(grid[i][2-i] for i in range(3))
    
    if ('X' not in diagon1 or 'O' not in diagon1) and ' ' not in diagon1:
        if diagon1[0] == player:
            return True, 1
        else:
            return True, -1
        
    elif ' ' in diagon1:
            end = False
            
    if ('X' not in diagon2 or 'O' not in diagon2) and ' ' not in diagon2:
        if diagon2[0] == player:
            return True, 1
        else:
            return True, -1
        
    elif ' ' in diagon2:
            end = False
    
    return end, 0
        
```

```python
def compute_move(grid, computer, enemy):
    value_list = []
    play_list = []
    
    for i in range(3):
        for j in range(3):
            if grid[i][j] == ' ':

                grid[i][j] = computer
                play_list.append([i, j])
                if evaluate_game(grid, computer)[0]:
                    value_list.append(evaluate_game(grid, computer)[1])
                else:
                    value_list.append(-compute_move(grid, enemy, computer)[0])
                    
                grid[i][j] = ' '
    return max(value_list), play_list[value_list.index(max(value_list))]
```

```python
#LITTLE FULL GAME, Player goes first
game = [[' ', ' ', ' '], [' ', ' ', ' '], [' ', ' ', ' ']]
loop = True
player = 'X'
computer = 'O'
printer(game)
print()
print('############################')
print()
while loop:
    if evaluate_game(game, player)[0]:
        loop = False
        if evaluate_game(game, player)[1] == 0:
            print('Empate. No es un juego muy dificil.')
            input()
        else:
            print('Derrota. Solo eres un patético humano a merced de la IA.')
            input()
    else:
        repeat = True
        while repeat:
            player_move = input('Escribe tu movimiento: ')
            [i, j] = [int(player_move[0]), int(player_move[1])]
            if game[i][j] == ' ':
                repeat = False
                game[i][j] = player
                printer(game)
                print()
                print('############################')
                print()
        if evaluate_game(game, player)[0]:
            loop = False
            if evaluate_game(game, player)[1] == 0:
                print('Empate. No es un juego muy dificil.')
                input()
            else:
                print('Derrota. Solo eres un patético humano a merced de la IA.')
                input()
        else:
            
            [res, [i, j]] = compute_move(game, computer, player)
            game[i][j] = computer
            printer(game)
            print()
            print('############################')
            print()
            if res==1 and not evaluate_game(game, player)[0]:
                print('La todopoderosa IA predice que va a derrotarte con un 100% de probabilidades.')
            
```
